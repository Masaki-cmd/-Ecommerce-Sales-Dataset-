````python
import zipfile
import os

zip_file_path = '/content/archive (1).zip'
extract_dir = '/content/extracted_data'

# Create the extraction directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f"'{zip_file_path}' extracted to '{extract_dir}'")

# List the contents of the extracted directory
print("Contents of extracted_data:")
for root, dirs, files in os.walk(extract_dir):
    for name in files:
        print(os.path.join(root, name))
    for name in dirs:
        print(os.path.join(root, name))
````

````python
import pandas as pd

details_df = pd.read_csv('/content/extracted_data/Details.csv')
orders_df = pd.read_csv('/content/extracted_data/Orders.csv')

print("Details.csv Head:")
print(details_df.head())
print("\nDetails.csv Info:")
details_df.info()

print("\nOrders.csv Head:")
print(orders_df.head())
print("\nOrders.csv Info:")
orders_df.info()
````

````python
# Convert 'Order Date' to datetime
orders_df['Order Date'] = pd.to_datetime(orders_df['Order Date'], format='%d-%m-%Y')

# Check for duplicates in both dataframes
print("\nDuplicates in details_df:", details_df.duplicated().sum())
print("Duplicates in orders_df:", orders_df.duplicated().sum())

# Check common columns for merging
common_cols = list(set(details_df.columns) & set(orders_df.columns))
print("\nCommon columns for merging:", common_cols)

# Display info again to confirm datetime conversion
print("\nOrders.csv Info after date conversion:")
orders_df.info()
````

````python
merged_df = pd.merge(details_df, orders_df, on='Order ID', how='inner')

print("Merged DataFrame Head:")
print(merged_df.head())

print("\nMerged DataFrame Info:")
merged_df.info()
````

````python
# カテゴリごとの売上と利益を計算
category_performance = merged_df.groupby('Category').agg(
    Total_Amount=('Amount', 'sum'),
    Total_Profit=('Profit', 'sum')
).sort_values(by='Total_Amount', ascending=False)

print("\nカテゴリごとの売上と利益:\n")
print(category_performance)

# サブカテゴリごとの売上と利益を計算
subcategory_performance = merged_df.groupby('Sub-Category').agg(
    Total_Amount=('Amount', 'sum'),
    Total_Profit=('Profit', 'sum')
).sort_values(by='Total_Amount', ascending=False)

print("\nサブカテゴリごとの売上と利益:\n")
print(subcategory_performance)
````

## Business Understanding & EDA (Temporal Analysis)

````python
# 年ごとの売上と利益の傾向を分析
annual_performance = merged_df.groupby(merged_df['Order Date'].dt.year).agg(
    Total_Amount=('Amount', 'sum'),
    Total_Profit=('Profit', 'sum')
).reset_index()
annual_performance = annual_performance.rename(columns={'Order Date': 'Year'})

print("\n年ごとの売上と利益:\n")
print(annual_performance)

# 月ごとの売上と利益の傾向を分析（全期間の月平均）
monthly_performance = merged_df.groupby(merged_df['Order Date'].dt.month).agg(
    Total_Amount=('Amount', 'sum'),
    Total_Profit=('Profit', 'sum')
).reset_index()
monthly_performance = monthly_performance.rename(columns={'Order Date': 'Month'})

print("\n月ごとの売上と利益:\n")
print(monthly_performance)
````

## Detect Outliers

### Subtask:
Detect and analyze outliers in numerical columns (`Amount`, `Profit`, `Quantity`) within the `merged_df` using a statistical method such as the Interquartile Range (IQR). Report the identified outliers, their potential impact, and any proposed handling strategy (though no actual removal/modification is required at this stage unless explicitly instructed).


**Reasoning**:
I will calculate the IQR-based outliers for 'Amount', 'Profit', and 'Quantity' columns, print the count of outliers, and display the first few rows of outlier data for each column as per the instructions.



````python
import numpy as np

numerical_cols = ['Amount', 'Profit', 'Quantity']
outliers_dict = {}

for col in numerical_cols:
    Q1 = merged_df[col].quantile(0.25)
    Q3 = merged_df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = merged_df[(merged_df[col] < lower_bound) | (merged_df[col] > upper_bound)]
    outliers_dict[col] = outliers

    print(f"\n--- Outliers for '{col}' ---")
    print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
    print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    print(f"Number of outliers: {len(outliers)}")
    if not outliers.empty:
        print("Sample Outliers (first 5 rows):")
        print(outliers.head())
    else:
        print("No outliers found for this column.")

````
